# 3. 데이터 품질 점검

```python
# 1. 이미지 무결성 확인 
broken_files = []
size_list = []

for parent_folder_name in os.listdir(src_dir):
    for child_folder_name in os.listdir(f"{src_dir}/{parent_folder_name}/"):
        for fname in os.listdir(f"{src_dir}/{parent_folder_name}/{child_folder_name}/img"):
              try:
                     img = Image.open(f"{src_dir}/{parent_folder_name}/{child_folder_name}/img/" + fname)
                     size_list.append(f'({img.size[0]}, {img.size[1]})')
                     img.verify()
              except:
                     broken_files.append(fname)
print("깨진 파일:", broken_files)
print("이미지 해상도 목록:")
print(np.unique(size_list, return_counts=True))
```

```python
# 2. 비디오 무결성 검증?
def is_video_broken_opencv(video_path):
    try:
        cap = cv2.VideoCapture(video_path)
        if not cap.isOpened():
            return True  # 파일 열기 실패

        while True:
            ret, frame = cap.read()
            if not ret:
                break  # 파일 끝 또는 오류
    except Exception as e:
        print(f"Error processing video: {e}")
        return True  # 예외 발생 시 깨진 것으로 간주
    finally:
        if 'cap' in locals() and cap.isOpened():
            cap.release()
    return False
```

```python
video_file = f"{src_dir}/SGA2100300/SGA2100300S0001/video/SGA2100300S0001.mp4"
if is_video_broken_opencv(video_file):
    print(f"{video_file} is broken.")
else:
    print(f"{video_file} is not broken.")
```

```python
broken_videos = []
for parent_folder_name in os.listdir(src_dir):
    for child_folder_name in os.listdir(f"{src_dir}/{parent_folder_name}/"):
        for fname in os.listdir(f"{src_dir}/{parent_folder_name}/{child_folder_name}/video"):
              try:
                     is_video_broken_opencv(f"{src_dir}/{parent_folder_name}/{child_folder_name}/video/"+fname)
              except:
                     broken_videos.append(fname)
```
